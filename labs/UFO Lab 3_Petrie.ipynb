```python
import requests
from bs4 import BeautifulSoup
```

```python
months = ['201501', '201502', '201503', '201504', '201505', '201506', '201507', '201508', '201509', '201510', '201511', '201512']
all_rows = []
for month in months:
    url = 'http://www.nuforc.org/webreports/ndxe' + month + '.html'
    response = requests.get(url)
    print response.text
    parser = BeautifulSoup(response.text, 'html.parser')
    all_rows = all_rows + parser.findAll('tr', attrs = {'valign':'TOP'})
```

```python
ufo_sightings = {
    'date':[],
    'city':[],
    'state':[],
    'shape':[],
    'duration':[],
    'summary':[],
    'date_posted':[],
}

for row in all_rows:
    ufo_info = []
    for cell in row.findChildren():
        ufo_info.append(cell.text)
    ufo_sightings['date'].append(ufo_info[0])
    ufo_sightings['city'].append(ufo_info[3])
    ufo_sightings['state'].append(ufo_info[5])
    ufo_sightings['shape'].append(ufo_info[7])
    ufo_sightings['duration'].append(ufo_info[9])
    ufo_sightings['summary'].append(ufo_info[11])
    ufo_sightings['date_posted'].append(ufo_info[13])
    
ufo_sightings
```
